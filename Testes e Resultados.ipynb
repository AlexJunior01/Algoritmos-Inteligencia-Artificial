```python
# Importando bibliotecas utilizadas
from ipynb.fs.defs.Singlelink import single_link
from ipynb.fs.defs.kmedias import kmedias

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import adjusted_rand_score
```

```python
# Lendo os conjuntos de dados
df1 = pd.read_csv('datasets/c2ds1-2sp.txt', sep='\t', index_col='sample_label')

df2 = pd.read_csv('datasets/c2ds3-2g.txt', sep='\t', index_col='sample_label')

df3 = pd.read_csv('datasets/monkey.txt', sep='\t', index_col='sample_label')
df3.columns = ['d1', 'd2']
```

```python
cores = {0: "violet", 1: "palegreen", 2: "lightgray",
        3: "rosybrown", 4: "khaki", 5: "cornflowerblue", 6: "coral", 7: "black"}
```

```python
# Lendo os resultados para serem comparados
dfresult1 = pd.read_csv('datasets/c2ds1-2spReal.clu', sep='\t', header=None)
dfresult1.columns = ["sample_label", "clusters"]
dfresult1.index = dfresult1["sample_label"]
dfresult1.drop(labels=["sample_label"], axis="columns", inplace=True) 

dfresult2 = pd.read_csv('datasets/c2ds3-2gReal.clu', sep='\t', header=None)
dfresult2.columns = ["sample_label", "clusters"]
dfresult2.index = dfresult2["sample_label"]
dfresult2.drop(labels=["sample_label"], axis="columns", inplace=True) 

dfresult3 = pd.read_csv('datasets/monkeyReal1.clu', sep='\t', header=None)
dfresult3.columns = ["sample_label", "clusters"]
dfresult3.index = dfresult3["sample_label"]
dfresult3.drop(labels=["sample_label"], axis="columns", inplace=True) 
```

# Resultados

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])

ax.set_title("Kmedias - Dataset 1", fontsize = 20)

for i in range(len(centroides_r1_k2)):
        temp = df_kmedias_r1_k2.query(f'clusters == {i}')
        plt.scatter(x=temp.d1, y=temp.d2, color = cores[i])
        plt.scatter(x=centroides_r1_k2[i][0], y=centroides_r1_k2[i][1],
                   marker='o', s=500, color = cores[i])
        
plt.show()
```
